# SERIES COMPOSITION AND SPARSE VECTOR

One of the most important properties of DP mechanisms is the composability: if n DP mechanisms each with $\varepsilon_1, \varepsilon_2, ..., \varepsilon_n$ privacy budgeds are applied to the data (either in series or in parallel), we obtain a DP mechanism.

If the mechanisms are applied in series, the new mechanism $\varepsilon$ is $\varepsilon =\sum^n_{i=1}\varepsilon_i$


If the mechanisms are applied in parallel, the new mechanism $\varepsilon$ is $\varepsilon =\max\varepsilon_i$
